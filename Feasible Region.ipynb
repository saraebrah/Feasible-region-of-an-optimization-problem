## Plotting the intersection of hyperplanes

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
plt.rcParams.update(plt.rcParamsDefault)
matplotlib.rcParams['mathtext.fontset'] = 'stix'
matplotlib.rcParams['font.family'] = 'STIXGeneral'
```

```python
plt.figure(figsize=(4.5,5.5))


# Plotting the feasible points   
xx = [1,  2,  4,  4,  3,  2,  3,  3,  4,  1,  2,    1,  5,    5,  2.7,  2.3,  1.4,  3.8,  4.8]
yy = [1,  1,  2,  5,  6,  4,  4,  2,  3,  3,  2.5,  5,  2.5,  4,  3.2,  4.7,  4.8,  4.3,  3.3]
plt.scatter(xx, yy, color="#008000",s=10, zorder=10)
plt.scatter([1], [1], color="#DC143C",s=10, zorder=10)
plt.text(0.55,0.55,'$X^0$',size=10, color='#DC143C')


# Plotting the cost vector
plt.arrow(5.5,5.5, 0.3, 0.3, head_width=0.1, head_length=0.1,linewidth=0.3,color="#000000")
plt.plot([5.3,5.7],[5.7,5.3],linewidth=0.4, color="#000000")
plt.text(5.7,5.5,'$c$',size=8, color='#000000')


# Plotting the hyperplanes
x = np.arange(0,11,0.1)
y0 = -x + 2
y1 = - x + 9
y2 = 0.5*x
y3 = 0.5*x + 4.5
plt.plot(x, y0,linestyle='dotted',linewidth=1.5, color="#DC143C")
plt.axvline(5, 0, 8,linestyle='dotted',linewidth=1.5, color="#DC143C")
plt.plot(x, y1,linestyle='--',linewidth=1.2, color='#808080')
plt.plot(x, y2,linestyle='--',linewidth=1.2, color='#808080')
plt.plot(x, y3,linestyle='--',linewidth=1.2, color='#808080')
plt.axvline(1, 0, 8,linestyle='--',linewidth=1.2, color='#808080')
plt.axvline(5, 0, 8,linestyle='--',linewidth=1.2, color='#808080')
plt.axhline(1, 0, 7,linestyle='--',linewidth=1.2, color='#808080')


# Finding the intersection of hyperplanes and shading the area in between (clockwise)
corner_x = [1,1,3,5,5,2]
corner_y = [1,5,6,4,2.5,1]
plt.fill(corner_x, corner_y,alpha=0.5, facecolor="none", hatch="////", edgecolor="#FFFF00" )


plt.xlim(0, 6.5)
plt.xticks(range(1,7))
plt.ylim(0, 8.5)
plt.grid(True,alpha=0.1)
plt.show()
```

```python
plt.figure(figsize=(5,5))


# Removing the top and right spines
plt.gca().spines['right'].set_color('none')
plt.gca().spines['top'].set_color('none')
plt.gca().xaxis.set_ticks_position('bottom')
plt.gca().spines['bottom'].set_position(('data',0))
plt.gca().yaxis.set_ticks_position('left')
plt.gca().spines['left'].set_position(('data',0))
plt.gca().plot(1, 0, ">k", transform=plt.gca().get_yaxis_transform(), clip_on=False)
plt.gca().plot(0, 1, "^k", transform=plt.gca().get_xaxis_transform(), clip_on=False)

# Hiding the axes
plt.gca().get_yaxis().set_visible(False) 
plt.gca().get_xaxis().set_visible(False)

# Plotting the feasible points      
xx = [2,2,4,4,3,3]
yy = [2,4,4,2,6,3]
plt.scatter(xx, yy, color="#000000",s=20, zorder=10)
plt.scatter([3], [6], color="#DC143C",s=15, zorder=10) # Highlighting the optimal solution in red
plt.text(3.1,6.2,'$X^0$',size=11, color='#DC143C') 


# Plotting the cost vector
plt.arrow(5,5, -0.2, -0.3, head_width=0.1, head_length=0.1,linewidth=0.5,color="#000000") # 
plt.plot([4.9,5.1],[5.1,4.9],linewidth=0.6, color="#000000")
plt.text(5,4.5,'$c$',size=11, color='#000000')


# Plotting the hyperplanes
x0 = np.arange(-1,10,0.1)
x1 = np.arange(2,5.5,0.1)
x2 = np.arange(1,2.1,0.1)
y0 = -x1 + 9
y1 = -x0 + 9
y2 = -x2 + 5
y3 = -1.5*x0 + 5

plt.plot(x1, y0,linestyle='dotted',linewidth=1.7, color="#DC143C", label= "Prior constraints",zorder=10)
plt.plot(x2, y2,linestyle='dotted',linewidth=1.7, color="#DC143C",zorder=10)
plt.plot([1,1], [4,7],linestyle='dotted',linewidth=1.7, color="#DC143C",zorder=10)
plt.plot([1,2], [7,7],linestyle='dotted',linewidth=1.7, color="#DC143C",zorder=10)
plt.plot([5.5,5.5], [3,3.5],linestyle='dotted',linewidth=1.7, color="#DC143C",zorder=10)
plt.plot([2,5.5], [3,3],linestyle='dotted',linewidth=1.7, color="#DC143C",zorder=10)

plt.plot(x0, y1,linestyle='--',linewidth=1.3, color="#808080", label= "Inferred constraints")
plt.plot(x0, y3,linestyle='--',linewidth=1.3, color="#808080")
plt.axvline(1, 0, 1,linestyle='--',linewidth=1.3, color="#808080")
plt.axhline(2, 0, 1,linestyle='--',linewidth=1.3, color="#808080")
plt.axvline(5.5, 0, 1,linestyle='--',linewidth=1.3, color="#808080")
plt.axhline(7, 0, 1,linestyle='--',linewidth=1.3, color='#808080')


# Finding the intersection of hyperplanes and shading the area in between (clockwise)
corner_x = [2,1,1,2,5.5,5.5,]
corner_y = [2,3.5,7,7,3.5,2,]
plt.fill(corner_x, corner_y,alpha=0.5, facecolor="none", hatch="//////", edgecolor="#A9A9A9" )


plt.xlim(-0.5, 7.3)
plt.ylim(-0.5, 8.3)
plt.grid(True,alpha=0.1)
plt.legend(loc='upper center', bbox_to_anchor=(0.6, -0.02),ncol=2,fontsize=10)
plt.show()
```
